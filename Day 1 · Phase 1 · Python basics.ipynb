# 1 Variables & assignment

what is Variables 

A variable is a named container that stores a value in memory. You create one using = (the assignment operator). Python figures out the type automatically — no need to declare it.

```python
#create variables

Name="Surya"
Age=22
Height=5.9
is_student=False
```

```python
#Re-Assign Anytime
Age=23
```

Variable names: use lowercase, underscores for spaces (snake_case). No spaces, no starting with a number. 
    
---->Examples: my_name, total_price, age1

Multiple Assignment

Python lets you assign many variables on one line, or give multiple variables the same value.

```python
#assign Multiple at once

x,y,z=10,20,30
```

```python
#same value to many

a=b=c=0
```

```python
#swap two variables

x,y=y,x
```

Type conversion (casting)

You can convert between types using built-in functions. This is very common when reading user input (which is always a string) and you need a number

```python
# int to float

x = 10
y = float(x)
print(type(y))
print("Integer to Float :", y)
```

```python
#float to int

x = 9.8
y = int(x)

print(type(y))
print("Float to Integer :",y)
```

```python
#str to int

x = "100"
y = int(x)

print(type(x))
print("String to Integer :",y)
```

```python
#int to str

x = 25
age =str(x)

print(type(age))
print("Integer to String :",age)
```

```python
#str to float
x = "5.9"
height=float(x)

print(type(height))
print("String to Float :",height)
```

```python
#float to str

x = 5.9
height = str(x)

print(type(height))
print("Float to String :",height)
```

# 3 print()

print()- display output

print() displays values to the screen. You can print multiple things separated by commas, and control the separator and ending characters.

```python
# Basic print

print("hello,World!")
```

```python
# print a variable

name = "surya"
print("Name :",name)
```

```python
# print multiple values

print("Name :",name,"age :",23)
```

```python
# f-string (most common way)

print(f"my name is{name} and i am {22} year old")
```

```python
# sep and end parameters

print("a","b","c",sep="-")
print("Hello", end=" 👋")
```

f - strings - the best way to format

f-strings (formatted string literals) let you embed variables and expressions directly inside a string. Always prefix with f. Used everywhere in real Python code.

```python
name = "surya"
age = 22
score = 95.678
```

```python
# embed variable

print(f"name:{name}")
```

```python
# embed expression

print(f"Next year:{age + 1}")
```

```python
# format decimal places
 
print(f"Score:{score :.2f}")
```

```python
# uppercase

print(f"{name.upper()}")
```

# 4.input()

input() - get user input

input() pauses the program and waits for the user to type something and press Enter. It always returns a str — so convert it if you need a number.

```python
# basic input

name=input("Enter your name:")
print(f"Hello,{name}!")
```

```python
# input always returns string!

age = input("Enter your age:")
print(type(age))
```

```python
# convert to int to use as number

age = int(input("Enter your age:"))
print(f"In 5 years: {age +5}")
```

# 5.Comments

single & multi - line Comments

Comments are notes for humans — Python ignores them completely. Use # for single-line. For multi-line, use triple quotes """ (also used for docstrings).

```python
# This is a single-line comment 


x = 10 

# inline comment after code # multi-line using multiple # # Line 1 # Line 2 # Line 3
""" This is a multi-line string often used as a comment block or a docstring (function description) """
```
